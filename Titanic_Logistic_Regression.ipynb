```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/logistic-regression/Social_Network_Ads.csv')
data.drop(columns = ['Gender', 'User ID'], axis = 1, inplace = True)
data.head()
```

```python
y = data.iloc[: , -1]
X = data.iloc[: , : -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.70, random_state = 100)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) #Standardized X, y
```

```python
from sklearn.linear_model import LogisticRegression?
lm = LogisticRegression(random_state = 6)
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
y_pred = y_pred[: , np.newaxis]
```

```python
ans = lm.score(X_test, y_test)
ans
```
